```python
import pandas as p
import numpy as n
import sklearn as s
import matplotlib.pyplot as plt
```

```python
data=p.read_csv("Coffee_Chain_Sales .csv")
```

```python
data
```

```python
newdata=data.drop(["Area Code","Cogs","DifferenceBetweenActualandTargetProfit","Date","Inventory Margin","Margin","Market_size","Market","Marketing","Product_line",],axis=1)
```

```python
newdata
```

```python
newdata1=newdata.drop(["Product_type","Product","State","Target_cogs","Target_margin","Profit","Target_profit","Type","Sales"],axis=1)
```

```python
newdata1
```

```python
x=newdata1.iloc[:,0].values.reshape(-1,1)
y=newdata1.iloc[:,1].values.reshape(-1,1)
```

```python
y
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as t
```

# Splitting The Dataset

```python
x_train,x_test,y_train,y_test=t(x,y,test_size=0.2,random_state=0)
```

# Checking the Train And Test

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
predictvalu=lm.predict(x_test)
```

```python
predictvalu
```

```python
compa=p.DataFrame(x_test,columns=['CGPA'])
```

```python
compa['AVLDATA']=y_test
```

```python
compa["predictdata"]=predictvalu
```

```python
compa
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,predictvalu)
```

```python
mean_absolute_error(y_test,predictvalu)
```

```python
r2_score(y_test,predictvalu)
```
